```python
# Lambda function

def multiply_these(x,y):
  product = x*y
  print(product)
  return product

multiply_these(-2,9)

var1, var2 = 2,3
multiply_these(var1,var2)


multi_these = lambda x,y: x * y
multi_these(5,6)
```

```python
iris['speal_len_sq'] = iris.sepal_length.apply(lambda x: x**2)
iris.head(5)

iris['sepal_len_wid_avg'] = iris[['sepal_length', 'sepal_width']].apply(lambda x: (x.sepal_length+x.sepal_width)/2, axis=1)
iris.head()
# parsed out the length and width from the list by x.len/wid

```

```python
import numpy as np
import pandas as pd

sales = pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/sales.csv')
sales.head(2)

sales['Revenue'].sum()
sales.dtypes

sales.head(20)
# remove $ from Revenue columns
sales['Revenue'] = sales['Revenue'].str.strip('$')
sales['Revenue'] = sales['Revenue'].astype('int')
sales
# verify that Revenue is now an integer
assert sales['Revenue'].dtype == 'int'
```

```python
# making categories of data

marriage_status = pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/marriage_status.csv')
marriage_status.head(3)

marriage_status['marriage_status'].describe()

marriage_status['marriage_status'] = marriage_status['marriage_status'].astype('category')
marriage_status.describe()

# category gives index to each value type in the column, runs more efficiently through large datasets
```
