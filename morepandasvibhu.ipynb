```python
# Data Cleaning

# Pandas has data cleaning tools
import pandas as pd
dupes = pd.read_csv('./dupe-vals.csv')

dupes.tail(40)
len(dupes)

# Working with Duplicate Data
dupes.duplicated()
# identify duplicate rows and columns within a DataFrame

# How many rows are duplicate?
print(dupes.duplicated().sum())
# Remove the duplicate rows -- but be sure to keep one copy!
dupes = dupes.drop_duplicates()
len(dupes)

# Working with Missing Data
# Pandas primarily uses the data type np.nan from NumPy to represent missing data

import numpy as np
df_miss = pd.DataFrame({
    'x': [2,np.nan,1],
    'y': [np.nan, np.nan, 6],
    'z': [4, np.nan, np.nan]}
)
df_miss

# .dropna() drop all rows with any missing data in any column
df_drop_all = df_miss.dropna()
df_drop_all
# subset parameter takes a list of column names to specify which columns should have missing values (say drop row if na is in said column)
df_drop_x = df_miss.dropna(subset=['x'])
df_drop_x
# .fillna() replace missing values with whatever you set it to
df_filled = df_miss.fillna(df_miss.median())
df_filled
# filled the values in each column with the median value of the respective column



# Practice
verymessy = pd.read_csv('./very-messy-data.csv')

verymessy.head(30)
verymessy.duplicated()
print(verymessy.duplicated().sum())
len(verymessy)
verymessy = verymessy.drop_duplicates()
len(verymessy)

verymessy

# remove lines where missing species data
verymessy_species = verymessy.dropna(subset=['species'])
verymessy_species

# imputed data based on mean of the rest of the column
verymessy2 = verymessy.fillna(verymessy.mean())
verymessy2
```
