```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline


titanic_df = pd.read_excel('titanic3.xls', 'titanic3', index_col=None, na_values = ['NA'])

titanic_df.head()
titanic_df.describe()
titanic_df.drop(['ticket', 'cabin', 'boat', 'body'],axis=1).head()

titanic_df.isnull() # shows which values are n/a
titanic_df.isnull().sum() # aggregates of isnull()

pd.value_counts(titanic_df['survived']).plot.bar() # bar plot of those who survived vs. those who didn't
titanic_df['survived'].mean()
titanic_df.groupby(['sex','pclass'])['survived'].mean() # got means stratified by sex and pclass

titanic_df['sex'].count() # count of those with 'sex' designation
titanic_df[titanic_df['age']<18] # table with observations of those restricted to under 18 years of age


titanic_df.index # tells where indices begin (0) and end (1309)

```
